```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime  # objects that contain date and time
import pandas as pd # used for data analysis: read, write and re-arrange datasets
import random

eq_filename2020 = './PTS2020csv.csv'
data2020 = pd.read_csv(eq_filename2020)

# print(data2020)


date2020 = data2020.iloc[:,0]
pH2020 = data2020.iloc[:,1]
sal2020 = data2020.iloc[:,2]
temp2020 = data2020.iloc[:,3]

# print(Date)
# print(pH)
# print(Sal)
# print(Temp)

# convert to my datetime objects
dtimes2020 = np.empty_like(data2020.index, dtype=object)

for ii in data2020.index:
    dtimes2020[ii] = datetime.fromisoformat(date2020[ii])
    
start_time2020 = datetime.fromisoformat('2020-02-01')
end_time2020 = datetime.fromisoformat('2020-04-01')
 
eggs2020 = []
for i in range(len(sal2020)):
    n = random.randint(400,100000)
    eggs2020.append(n)
print(eggs2020)

maxeggs2020 = eggs2020[0]
index = 0
for i in range(1,len(eggs2020)):
    if eggs2020[i] > maxeggs2020:
        maxeggs2020 = eggs2020[i]
        index = i
        
print(maxeggs2020, index)

print(date2020[i])



########################################################################PLOT#######################################################################

fig, ((ax1, ax2, ax3, ax4)) = plt.subplots(4,1, figsize=(15,18))
plt.xlabel("Date (YYYY-MM-DD)", fontsize=16)
#plt.ylabel("Abiotic Factors (pH, Temperature, Salinity)", fontsize=16)

ax1.plot(dtimes2020, pH2020)
ax1.set_ylabel('pH')
ax1.set_xlim([start_time2020, end_time2020])
#ax1.axvline(x=date[i], color='r')
ax1.grid()

ax2.plot(dtimes2020, temp2020)
ax2.set_ylabel('Temperature (â„ƒ)')
ax2.set_xlim([start_time2020, end_time2020])
#ax2.axvline(x=date[i], color='r')
ax2.grid()

ax3.plot(dtimes2020, sal2020)
ax3.set_ylabel('Salinity (g/kg)')
ax3.set_xlim([start_time2020, end_time2020])
#ax3.axvline(x=date[i], color='r')
ax3.grid()

ax4.plot(dtimes2020, eggs2020)
ax4.set_ylabel('Egg Count (#)')
ax4.set_xlim([start_time2020, end_time2020])
#ax4.axvline(x=date[i], color='r')
ax4.grid()


plt.savefig('PTS2020.png', dpi=300, bbox_inches='tight')
```
